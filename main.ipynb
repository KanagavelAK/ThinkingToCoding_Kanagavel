```python
num = int(input("Enter a number "))
if num > 0:
  if num % 2 == 0:
    print("The number is Positive and Even")
  else:
    print("The number is Positive and Odd")
elif num < 0:
  if num % 2 == 0:
    print("The number is Negative and Even")
  else:
    print("The number is Negative and Odd")
else:
  print("The number is Zero")
```
